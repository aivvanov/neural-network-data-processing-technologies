Заметным образом ничего существенно не поменялось, есть лишь незначительные изменения некоторых показателях.

```python
# Выполнение нормализации атрибутов по трем методам
blood_pressure = df['BloodPressure']
blood_pressure.describe()
```

```python
# Нормализация по методу минимакса
minmax = (blood_pressure - blood_pressure.min()) / (blood_pressure.max() - blood_pressure.min())
minmax.describe()
```

```python
# Нормализация по Z-показателю
z_meth = (blood_pressure - blood_pressure.mean()) / blood_pressure.std()
z_meth.describe()
```

```python
# Десятичное масштабирование
mass = len(str(int(blood_pressure.max())))
dec = blood_pressure / 10**mass
dec.describe()
```

Анализируя результаты проведения нормализации по трем методам можно заметить, что диапазон значений метода десятичного масштабирования и нормализации по методу минимакса схож, хотя есть различия некоторых параметров на 1-ну 10-ую. Диапазон значений нормализации по Z-показателю сильно отличается от остальных методов.
