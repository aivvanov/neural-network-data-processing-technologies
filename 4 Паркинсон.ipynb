```python
# Определим переменные. В переменную «x» сохраните атрибуты из набора данных (признаки) 
x = df.loc[:, 'gender':'tqwt_kurtosisValue_dec_36']
```

```python
# В переменную «y» - значения из целевой колонки «class».
y = df['class']
```

```python
# Импортируем функцию train_test_split из модуля sklearn
from sklearn.model_selection import train_test_split
```

```python
# Разделяем данные из «х» и «у» на тренировочную и тестовую выборку.
split = train_test_split(x, y, test_size = 0.20, random_state = 42)
(x_train, x_test, y_train, y_test) = split
```

```python
# Проектирование модели нейронной сети. Для этого создадим класс с именем «Perceptron»
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense


class Perceptron:
    @staticmethod
    def build(dim, num_classes):
        model = Sequential()
        model.add(Dense(128, input_dim = dim, activation = 'relu'))
        model.add(Dense(32, activation = 'relu'))
        model.add(Dense(num_classes, activation = 'softmax'))
        return model
```

```python
x_train.shape[1]
```

```python
# Создание экземпляра класса «Perceptron»
model = Perceptron.build(x_train.shape[1], 2)
```

```python
# Компиляция созданной модели нейронной сети
model.compile(loss = 'sparse_categorical_crossentropy', optimizer = 'adam', metrics =['accuracy'])
print("[INFO] Model compiled")
```

```python
# Процесс обучения модели НС
H = model.fit(x_train, y_train, batch_size = 32, epochs = 100, verbose = 1, validation_split = 0.20)
```

```python
H_eval = model.evaluate(x_test, y_test, batch_size = 151)
```

```python
import matplotlib.pyplot as plt
```

```python
def plot(results):
    plt.plot(results.history['accuracy'], label = 'Аккуратность на обучающем наборе')
    plt.plot(results.history['val_accuracy'], label = 'Аккуратность на проверочном наборе')
    plt.xlabel('Эпоха обучения')
    plt.ylabel('Аккуратность')
    plt.legend()
    plt.show()
```

```python
plot(H)
```

```python
# Изменение архитектуры сети скрытых слоев на 256-128-64-32
class Perceptron:
    @staticmethod
    def build(dim, num_classes):
        model = Sequential()
        model.add(Dense(256, input_dim = dim, activation = 'relu'))
        model.add(Dense(128, activation = 'relu'))
        model.add(Dense(64, activation = 'relu'))
        model.add(Dense(32, activation = 'relu'))
        model.add(Dense(num_classes, activation = 'softmax'))
        return model
```
