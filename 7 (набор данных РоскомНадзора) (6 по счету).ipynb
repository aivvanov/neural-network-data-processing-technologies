### Произведите токенизацию текста для каждой записи

```python
import re
import string
from nltk.tokenize import word_tokenize 
```

```python
# Произведем токенизацию исходного набора для каждой записи. 
# Для разделения текста на отдельные слова и знаки воспользуемся метод word_tokenize().
# Создадим функцию clean_text при помощи инструкции def и построчно заменим символы, 
# от которых необходимо избавиться, на пустые значения.
def clean_text(tweet):
    tweet.lower()
    tweet = re.sub(r"http\S+|www\S+|http\S+", '', tweet, flags = re.MULTILINE)
    tweet = re.sub(r"\@\w+|\#", '', tweet)
    tweet = re.sub(r"MSK", '', tweet)
    tweet = re.sub(r"roscomnadzor", '', tweet)
    tweet = re.sub(r"[0-9]", '', tweet)
    tweet = tweet.translate(str.maketrans('', '', string.punctuation))

    tweet_tokens = word_tokenize(tweet)
    filtred_words = [w for w in tweet_tokens]
    return ' '.join(filtred_words)
```

```python
# Применяем функцию clean_text для нашего атрибута в датасете
df.Information = df.Information.apply(clean_text)
```

```python
# Повторно выведем первую строку атрибута Information
df.Information[0]
```

### Проведите стемминг исходных слов и измерьте время выполнения операции. Сохраните результат стемминга в переменную. 

```python
# Создаем новую переменную stem_df и дадим название его единственного атрибута 'tweets'
stem_df = pd.DataFrame(columns = ['tweets'])
stem_df
```

```python
from nltk.stem.snowball import SnowballStemmer
from datetime import timedelta
import time
```

```python
# Создадим функцию stemming_text при помощи инструкции def.
# Лемматизация использует словарь и морфологический анализ, 
# чтобы в итоге привести слово к его канонической форме – лемме.  
def stemming_text(tweet):
    tweet_tokens = word_tokenize(tweet)
    filtred_words = [w for w in tweet_tokens]
    snow_lem = SnowballStemmer('russian')
    stemmed_words = [snow_lem.stem(w) for w in filtred_words]
    return ' '.join(stemmed_words)
```

```python
# Измерим время выполнения операции.
start_time = time.monotonic()
stem_df.tweets = df.Information.apply(stemming_text)
end_time = time.monotonic()
print(timedelta(seconds = end_time - start_time))
```

```python
stem_df.tweets[0]
```

По полученному результату можно сделать вывод, что стемминг исходных слов набора данных произведен успешно.
В ходе стемминга происходит отсечение частей слова, таким образом, что от слова остается только корень, что и можно наблюдать в ячейке [28].

### Проведите лемматизацию исходных слов и измерьте время выполнения операции. Сравните время выполнения стемминга и лемматизации. 

Лемматизация – процесс, который использует словарь и морфологический анализ, чтобы в итоге привести слово к его канонической форме – лемме. Для лемматизации русских и украинских текстов используется библиотека pymorphy2. 
Метод normal_forms() из класса MorphAnalyzer() позволяет нормализовать слово, приведенное в параметрах.

```python
!pip install pymorphy2
```

```python
lemmed_df = pd.DataFrame(columns = ['lem_tweets'])
lemmed_df
```

```python
import pymorphy2

def lemming_text(tweet):
    tweet_tokens = word_tokenize(tweet)
    filtred_words = [w for w in tweet_tokens]
    lemmatizer = pymorphy2.MorphAnalyzer()
    lemmed_words = [lemmatizer.normal_forms(w)[0] for w in filtred_words]
    return ' '.join(lemmed_words)
```

```python
# Измерим время выполнения операции.
start_time = time.monotonic()
lemmed_df.lem_tweets = df.Information.apply(lemming_text)
end_time = time.monotonic()
print(timedelta(seconds = end_time - start_time))
```

```python
lemmed_df.lem_tweets[0]
```

Процесс выполенния операции леммитизации занимает примерно в два раза больше времени.
