```python
# Произведем замену категориальных значений на челочисленные
symbols = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'k', 'l', 'm', 'n', 'o', 'p', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
len(symbols)
#df.classes.replace({'e': 0, 'p':1}, inplace = True)
```

```python
# Для этого создадим пустой массив и цикл 
values = []
for i in range(22):
    values.append(i)
print(values)
print(len(values))
```

```python
# Произведем замену значений массива symbols на values
df2 = df.replace(to_replace=symbols, value=values)
df2
```

По полученному результату видно, что замена произведена корректно.

#  Выполните преобразование любого атрибута по методу Label Encoding. Какой существенный минус есть у данного метода?

Label Encoding – сопоставление каждого значения категории с некоторым числом. Описанная операция может быть выполнена с помощью метода map(), который имеет следующие параметры

```python
# Заменим исходные значения аргумента gill-attachment с катег. на целочисл. используя данный метод
ga_new={'f':5, 'a': 0}
df['gill-attachment']=df['gill-attachment'].map(ga_new)
```

```python
df
```

Как можно видеть в результате атрибут gill-attachment теперь имеет целочл. значения - замена прошла успешно.
Минусом метода Label Encoding можно считать, что аргументы могут принимать значения от 0 до + бесконечности.

```python
from sklearn.preprocessing import LabelEncoder
encod = LabelEncoder()
```

```python
encod.fit_transform(df.classes)
```

# Выполните преобразование всех атрибутов по методу One-Hot Encoding. Выведите получившийся результат. Какой минус есть у данного метода?

Метод One-Hot Encoding – исходный атрибут преобразуется в бинарный признак. При замене данным методом для каждой категории атрибута создается отдельный столбец, поля которого могут принимать значение 0, когда исходное значение не принадлежит данному классу, и 1 – когда принадлежит. 

```python
# Заменим исходные значения аргументов по бинарному признаку, используя данный метод
df4 = pd.get_dummies(df, columns = ['classes', 'cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor',
       'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color',
       'stalk-shape', 'stalk-root', 'stalk-surface-above-ring',
       'stalk-surface-below-ring', 'stalk-color-above-ring',
       'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number',
       'ring-type', 'spore-print-color', 'population', 'habitat'])
```

```python
df4
```

```python
df4.describe()
```

Метод Метод One-Hot Encoding можно использовать для определения принадлежности конкретной категории к классу (0 или 1).
Минус: большие объемы информации после преобразования данных, количетство столбцов увеличелось с 23 до 119

Вывод: Метод Label Encoding лучше использовать для частного преобразования аргуметов, нежели при замене типа данных всего датасета целиком. В этом случае будет целесообразно использовать Метод One-Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
encdr = OneHotEncoder(sparse = False)
```

```python
encdr.fit_transform(df[['cap-shape']])
```
