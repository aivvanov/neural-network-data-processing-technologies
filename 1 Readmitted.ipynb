```python
def pie(column, df = df,  text = "Graph"):
    x = []
    unique = df[pd.notna(df[column])][column].unique()
    unique.sort()
    for label in unique:
        x.append(df[df[column]==label][column].count())
    
    fig = plt.figure(figsize=(6,6))
    patches, _ = plt.pie(x, labels=x)
    plt.title(text)
    plt.legend(patches, unique, loc='upper left')
    plt.show()
```

```python
# определение влияния max_glu_serum на readmitted
df['max_glu_serum'].unique()
# Norm = 0, >200 = 1, >300 = 2 <---- значение max_glu_serum
df.loc[df['max_glu_serum']=='Norm',['max_glu_serum']] = 0
df.loc[df['max_glu_serum']=='>200', ['max_glu_serum']] = 1
df.loc[df['max_glu_serum']=='>300', ['max_glu_serum']] = 2
max_glu = []
readmitted = []
glu_count = []
for j in df[pd.notna(df['max_glu_serum'])]['readmitted'].unique():
    for i in [0,1,2]:
        count = df[(df['max_glu_serum'] == i) & (df['readmitted'] == j)]['readmitted'].count()
        max = df[(df['max_glu_serum'] == i)]['readmitted'].count()
        glu_count.append((j,count/max*100))
fig, ax = plt.subplots()
x = np.arange(3)
width = 0.2
ax.bar(x, [x[1] for x in glu_count[:3]], width, color='#00AA22', label=glu_count[0][0])
ax.bar(x+width, [x[1] for x in glu_count[3:6]], width, color='#22AABB', label=glu_count[3][0])
ax.bar(x+width*2, [x[1] for x in glu_count[6:]], width, color='#4466EE', label=glu_count[6][0])
ax.set_xticks(x + width)
ax.set_xticklabels(['Norm','>200','>300'])
ax.legend()
plt.show()
```

Зависимость readmitted от Max_Glu_Serum незначительна

```python
# определение влияния A1Cresult на readmitted
results = ['Norm', '>7', '>8']
readmitted_values = ['NO', '<30', '>30']
colors = ['#00AA22', '#22AABB', '#4466EE']
results_df = df[pd.notna(df['A1Cresult'])][['A1Cresult','readmitted']]
readmitted = []
result_count = []
for i in range(len(results)):
    results_df.loc[results_df['A1Cresult']==results[i],['A1Cresult']] = i
for j in readmitted_values:
    for i in range(len(results)):
        count = results_df[(results_df['A1Cresult'] == i) & (results_df['readmitted'] == j)]['readmitted'].count()
        max = results_df[(results_df['A1Cresult'] == i)]['readmitted'].count()
        result_count.append((j,count/max*100))
fig, ax = plt.subplots()
x = np.arange(3)
rslt_cnt = len(results)
width = 0.2
for i in [0,1,2]:
    x_ = x+width*i
    ax.bar(x_, [y[1] for y in result_count[0+i*rslt_cnt:rslt_cnt+i*rslt_cnt]], width, color=colors[i%3], label=result_count[i*rslt_cnt][0])
ax.set_xticks(x + width)
ax.set_xticklabels(results)
ax.legend()
plt.show()
```

Влияние A1Cresult(результат теста A1c) на значение readmitted несущетсвенное

```python
# определение влияния weigth на readmitted
weights = ['[0-25)','[25-50)', '[50-75)', '[75-100)', '[100-125)', '[125-150)', '[150-175)','[175-200)']
readmitted_values = ['NO', '<30', '>30']
colors = ['#00AA22', '#22AABB', '#4466EE']
weights_df = df[pd.notna(df['weight']) & df['weight'] != '>200'][['weight','readmitted']]
for i in range(len(weights)):
    weights_df.loc[weights_df['weight']==weights[i],['weight']] = i
weights_count = []
for j in readmitted_values:
    for i in range(len(weights)):
        count = weights_df[(weights_df['weight'] == i) & (weights_df['readmitted'] == j)]['readmitted'].count()
        max = weights_df[(weights_df['weight'] == i)]['readmitted'].count()
        weights_count.append((j,count/max*100))
fig, ax = plt.subplots(figsize=(14,3))
wgt_cnt = len(weights)
x = np.arange(8)
width = 0.2
for i in [0,1,2]:
    x_ = x+width*i
    ax.bar(x_, [y[1] for y in weights_count[0+i*wgt_cnt:wgt_cnt+i*wgt_cnt]], width, color=colors[i%3], label=weights_count[i*wgt_cnt][0])
ax.set_xticks(x + width)
ax.set_xticklabels(weights)

ax.legend()

plt.show()
```

влияние веса на значение readmitted несущетсвенное, если не учитывать вес 0-25 (слишком маленький)

Из графика видно, что более половины пациентов повторно больницу не посещали. Около 10% повторно посещали в течении месяца и около 30% повторных посещений спустя более одного меяца.

```python
# Взять один из влияющих на переменную readmitted атрибутов и выполнить группировку и подсчет 
levels = ['No', 'Steady', 'Up', 'Down']
readmitted_values = ['NO', '<30', '>30']
colors = ['#00AA22', '#22AABB', '#4466EE']
levels_df = df[pd.notna(df['insulin'])][['insulin','readmitted']]
for i in range(len(levels)):
    levels_df.loc[levels_df['insulin']==levels[i],['insulin']] = i
levels_count = []
for j in readmitted_values:
    for i in range(len(levels)):
        count = levels_df[(levels_df['insulin'] == i) & (levels_df['readmitted'] == j)]['readmitted'].count()
        max = levels_df[(levels_df['insulin'] == i)]['readmitted'].count()
        levels_count.append((j,count/max*100))
print(levels_count)
fig, ax = plt.subplots(figsize=(14,3))
lvl_cnt = len(levels)
x = np.arange(lvl_cnt)
width = 0.1
for i in [0,1,2]:
    x_ = x+width*i
    ax.bar(x_, [y[1] for y in levels_count[0+i*lvl_cnt:lvl_cnt+i*lvl_cnt]], width, color=colors[i%3], label=levels_count[i*lvl_cnt][0])
ax.set_xticks(x + width)
ax.set_xticklabels(levels)
ax.legend()
plt.show()
```

Исходя из полученных графиков, можно сделать вывод, что люди, принимавшие инсулин, и люди, которым он не был прописан, обращаются в польницу повторно в равной степени. Увеличение и снижение дозы инсулина также не приводит к уменьшению шанса на повторный прием пациента.
